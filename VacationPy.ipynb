```python
# Store 'Lat' and 'Lng' into  locations 


# Convert Poverty Rate to float and store
# HINT: be sure to handle NaN values
Vacay_df = Vacay_df.dropna()
Humids = Vacay_df["Humidity"].astype(float)
locations = Vacay_df[["Latitude", "Longitude"]].astype(float)
```

```python
# Zero cloudiness.
Zero_Clouds = Vacay_df.loc[(Vacay_df["Clouds %"] == "0")]


# A max temperature lower than 80 degrees but higher than 70.
Zero_Clouds['Temperature_F'] = pd.to_numeric(Zero_Clouds['Temperature_F'])
Zero_Clouds['Wind Speed (mph)'] = pd.to_numeric(Zero_Clouds['Wind Speed (mph)'])
Nice_Temp = Zero_Clouds.loc[(Zero_Clouds["Temperature_F"]>70)]
Nice_Temp = Nice_Temp.loc[(Nice_Temp["Temperature_F"]< 80)]
Ideal_df = Nice_Temp.loc[Nice_Temp["Wind Speed (mph)"]<10]
# Wind speed less than 10 mph.

Ideal_df
```

```python

radius =5000

# Build URL using the Google Maps API
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
hotels =[]

for index, row in Ideal_df.iterrows():
    #city = row['City']
    lats = row["Latitude"]
    lngs = row["Longitude"]
   
    # redefine params
    params = {
        "location": f"{lats},{lngs}" ,
        "types": "hotel",
        "radius": radius,
        "key": g_key
    }

    # Run request
    balloons = requests.get(base_url, params=params).json()
    try:
        hotels.append(balloons['results'][0]['name'])
    except:
        hotels.append("")
        continue
        

Ideal_df['Hotel Name']=hotels
Ideal_df = Ideal_df.dropna(how='any')
Ideal_df
```

```python
Ideal_df["Latitude"] = pd.to_numeric(Ideal_df["Latitude"])
Ideal_df["Longitude"] = pd.to_numeric(Ideal_df["Longitude"])
locations2 = Ideal_df[["Latitude", "Longitude"]]
```
