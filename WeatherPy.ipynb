```python
# Dependencies
import matplotlib.pyplot as plt
import requests
from scipy import stats
import pandas as pd
from config import api_key
from citipy import citipy
import numpy as np
import random
```

```python
#output_data_file = "output_data/cities.csv"
```

```python
#How to get Random Cities? (Randomly select at least 500 unique (non-repeat) cities based on latitude and longitude.)
#500 random latitude and longitudes
Lats = np.random.uniform(-60,75,970)
Longs = np.random.uniform(-160,175,970)
coor = list(zip(Lats, Longs))
```

```python
cities=[]
for x in coor[0:]:
    print(x)
    city = citipy.nearest_city(x[0],x[1]).city_name
    if city not in cities:
        cities.append(city)
```

```python
print(len(cities))
```

```python
print(cities)
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
# #test response to get data out of json
response1 = requests.get(query_url + "semnan").json()
response1
```

```python
#build a dataframe to store data
#Divide Latitudes into N/S
#df to partition
#make dictionary
Weather_dict = {
    "City": cities ,
    "Latitude": "" ,
    "Longitude": "" ,
    "Temperature" : "",
    "Humidity": "",
    "Clouds %" : "",
    "Wind Speed (mph)": ""
    
}
Weat = pd.DataFrame(Weather_dict)
Weat
```

```python
# set up lists to hold reponse info
lat = []
temp = []
hum = []
clouds = []
winds = []

citynumber = 0
# Ref: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iterrows.html
# Loop through the list of cities and perform a request for data on each
for index, row in Weat.iterrows():
    city = row['City']
    response = requests.get(query_url + city).json()
    
    try:
        Weat.loc[index, "Latitude"] = response['coord']['lat']
        Weat.loc[index, "Longitude"] = response['coord']['lon']
        Weat.loc[index, "Temperature"] = response['main']['temp']
        Weat.loc[index, "Humidity"] = response['main']['humidity']
        Weat.loc[index, "Clouds %"] = response['clouds']['all']
        Weat.loc[index, "Wind Speed (mph)"] = response['wind']['speed']
        print(f"City number: {citynumber} - {city}")
        citynumber += 1
        
#         hum.append(response['main']['humidity'])
#         clouds.append(response['clouds']['all'])
#         winds.append(response['wind']['speed'])
        
 
      
        
    # Handle exceptions 
    except:
        # Append null values
        print("City not found")
        #continue?
        continue
```

```python
Weat['Temperature'] = pd.to_numeric(Weat['Temperature'])

#stole this function from here
## http://ucsdlib.github.io/win2016-python-gps/functions-pandas.html
def C_to_F(temp_celsius):
    temp_fahr = (temp_celsius *9/5)+32 
    return  temp_fahr

Weat["Temperature_F"] = C_to_F(Weat["Temperature"])
```

```python
#Northern Hemisphere Data
Northern_df = Weat.loc[(Weat["Latitude"] >= 0)]
Northern_df
```

```python
## Convert variables
Northern_df['Temperature'] = pd.to_numeric(Northern_df['Temperature'])
Northern_df['Latitude'] = pd.to_numeric(Northern_df['Latitude'])

#List variables
lat = Northern_df['Latitude']
temp = Northern_df['Temperature']

# Run linear regression
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(lat, temp)
regress_values = lat * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot scatter plot
plt.scatter(lat,temp)

# Plot regression line
plt.plot(lat,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")

# Label plot
plt.xlabel('Latitude')
plt.ylabel('Temperature')

# Print r square value
print(f"R squared: {rvalue**2}")

# Show plot
plt.show()
plt.savefig("NorthernTempVLatitude.png")
```

# Northern Hemisphere  Temp v Latitude
It gets warmer as you approach the equator. The R-squared gives us a strong correlation here.

```python
#Southern Hemisphere Data
Southern_df = Weat.loc[(Weat["Latitude"] < 0)]
Southern_df
```

```python
## Convert variables
Southern_df['Temperature'] = pd.to_numeric(Southern_df['Temperature'])
Southern_df['Latitude'] = pd.to_numeric(Southern_df['Latitude'])

#List variables
lat = Southern_df['Latitude']
temp = Southern_df['Temperature']

# Run linear regression
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(lat, temp)
regress_values = lat * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot scatter plot
plt.scatter(lat,temp)

# Plot regression line
plt.plot(lat,regress_values,"r-")
plt.annotate(line_eq,(-30,10),fontsize=15,color="red")

# Label plot
plt.xlabel('Latitude')
plt.ylabel('Temperature')

# Print r square value
print(f"R squared: {rvalue**2}")

# Show plot
plt.show()
plt.savefig("SouthernTempVLatitude.png")
```
